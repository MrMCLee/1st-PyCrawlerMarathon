```python
# 1. 取出班次一的每一個時間

'''
Your Code
'''
# 開啟 CSV 檔案
with open('./data/example.csv',encoding="utf-8", newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row[5])
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

'''
Your Code
'''
with open('./data/example.csv',encoding="utf-8", newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        time1 = str(row[5])
        print(f'Time : {time1}, type = {type(time1)}')
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

'''
Your Code
'''
with open('./data/example.csv',encoding="utf-8", newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        time = str(row[5])+'\t'+str(row[6])+'\t'+str(row[7])+'\t'+str(row[8])+'\t'+str(row[9])
        print(f'Time : {time}, type = {type(time)}')
```
