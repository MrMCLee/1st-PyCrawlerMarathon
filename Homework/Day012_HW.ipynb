### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d)
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    print('='*100)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
All_Tag_C = []
for AllClassification in soup.find(class_="part_list_2").find_all('em'):
    Tag_C = AllClassification.text
    
    if Tag_C in All_Tag_C:
        pass
    else:
        All_Tag_C.append(Tag_C)
print(f'Today total classification is\n{All_Tag_C}\nClassification is {len(All_Tag_C)}')
```

```python
artsclassification = []
for i in range(len(All_Tag_C)):
    Target_C = All_Tag_C[i]
    print(f'Classification is {Target_C}')
    print('-'*100)
    i = 0
    for d in soup.find(class_="part_list_2").find_all('h3'):
        art_c = d.find('em').text
        if art_c == Target_C:
            data = d.find(class_="date").text +' ' + d.find_all('a')[-1].text
            artsclassification.append(data)
            print(f'{data}')
            i=i+1
        else:pass
    print(f'Total arts is {i}')
    print('='*100)
#print(artsclassification, len(artsclassification))
```

### Use Pandas

```python
import pandas as pd

ettoday_arts = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    ettoday_arts.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])
    #print(d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text)

cols = ['Data', 'Class', 'Tital']    

ettoday_df = pd.DataFrame(ettoday_arts, columns=cols)
ettoday_df
```

```python
ettoday_classification = ettoday_df.groupby('Class')
ettoday_classification.groups
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
ettoday_df_max = pd.DataFrame(ettoday_df.groupby('Class').size(), columns=['Arts'])
ettoday_df_max.sort_values(by=['Arts'], ascending=False)
```
